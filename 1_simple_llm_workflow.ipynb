```python
from langgraph.graph import StateGraph,START,END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model=ChatOpenAI(model="gpt-3.5-turbo", temperature=0.9)
```

```python
#create a state
class LLMState(TypedDict): # type: ignore
    question: str
    answer: str
```

```python
def llm_qa(state: LLMState) -> LLMState:
    #extract the question
    question = state["question"]
    #add a prompt
    prompt = f"You are a helpful assistant. Answer the following question: {question}"
    #get the response from the model
    response = model.invoke(prompt)
    #update the state with the answer
    state["answer"] = response.content
    
    return state
```

```python
graph = StateGraph(LLMState)
graph.add_node("llm_qa",llm_qa)
graph.add_edge(START,"llm_qa")
graph.add_edge("llm_qa",END)
workflow = graph.compile()
```

```python
#execute workflow
initial_state = LLMState(question="What is the capital of France?", answer="")
result = workflow.invoke(initial_state)
print(result["answer"])
```
